### 打印出 100 以内所有的质数

质数的定义，大于 2 且只能被它自身和 1 整除的整数。

**算法**


```python
# 打印 100 以内所有的质数

for i in range(2, 100):
    if i == 2:
        print(i)
        continue
    for j in range(2, i):
        if (i%j) == 0:
            break
        else:
            print(i)
```

```python
for n in range(2, 101): #range(2,100)表示含左侧 2，不含右侧 100，是不是第三次看到这个说法了？
    if n == 2:
        print(n)
        continue
    for i in range(2, n):
        if (n % i) == 0:
            break
    else:                  # 这里目前你可能看不懂…… 但先关注结果吧。
        print(n)
```

## 语句块



在 python 语言中，**行首空白** （leading whitespace，由空格 ` ` 或者 Tab `⇥` 构成）有着特殊的含义。

如果有行首空白存在，那么，python 将认为这一行与其他临近的有着相同行首空白的语句同属于一个**语句块**——一个语句块必然由一个行末带有冒号 `:` 的语句起始。同属于一个语句块中的语句，行首空白数量应该相等。

```python
for i in range(10):
    if i % 2 != 0:
        print(i)
```

以上代码，一共包含三个语句，两个语句块。一个 `for` 循环语句块中包含一个 `if` 条件语句块。注意第一行和第二行的末尾的冒号 `:`。

在很多其他的语言中，比如，JavaScript，用大括号 `{}` 作为语句块标示。缩进是 python 比较特殊的地方。
