####这是一个数据主义的时代

终于找到一个可以社交，同时又学代码与英语，据说还有工作量证明的地方学习编程，这种形式很好，这是一条人类协作的大河，可以用代码交流，可以跨越国界、文化、历史 ，我们产生的信息实际上都可以记录到互联网上，未来不会开车没有关系，因为都用无人驾驶汽车了，未来是基因工程与计算机算法主导的世界，《未来简史》赫拉利说生物也是算法，这样看起来，对我来说，眼下重要的就是学点编程了，学这个并不是说我要靠编程为个人职业，而是希望通过自学，去发现一个不一样的自己，让未来人生的河流那怕转一个小弯，也是一种不错的经历。

####创意通过协作产生
今天不会编程的人就是文盲！**我今天就是文盲** ，但我相信通过社交、共享协作的办法，有可能让自己的大脑操作系统升一次级。学会了应该可以与更多的人一起进行原始性的协作，特别区块链领域的创意，应该用这种git的版本控制技术加上区块链技术结合起来，让创意的价值得到上链确权，所以学会这套方法就是创意协作的入门。
